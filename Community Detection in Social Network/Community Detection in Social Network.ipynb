## Community Detection in Social Network

Testing community package to find clusters of communities in social network graph from hamsterster.com.

About Dataset: Dataset taken from http://konect.uni-koblenz.de/networks/petster-friendships-hamster, dataset represent a Network containing friendships between users of the website hamsterster.com. Here Each Vertex represent a user & each Edge represent friendhip among 2 users.

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
#Reading dataset
interaction_graph = nx.read_edgelist("out.petster-friendships-hamster-uniq", create_using = nx.Graph(), nodetype = int)
```

```python
print(nx.info(interaction_graph))
```

_Its a grpah of 1858 users having 12534 connections or friendships, Lets plot this graph_

```python
plt.figure(figsize=(15,8))
spring_pos = nx.spring_layout(interaction_graph)
nx.draw_networkx(interaction_graph, pos = spring_pos, with_labels = False, node_size = 10)
```

### Finding Communities

Lets use community package now to find communities in this social network.

**Note** : Community detection results are speculative, they may not represent actual communities in the network, but a possible way in which communities may form in this network.

```python
import community
```

```python
plt.figure(figsize=(20,12))
part = community.best_partition(interaction_graph)
values = [part.get(node) for node in interaction_graph.nodes()]

nx.draw_spring(interaction_graph, cmap = plt.get_cmap('jet'), node_color = values, node_size=30, with_labels=False)
```
